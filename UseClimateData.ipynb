```python
def get_tmean_for_cluster_i(data,i):
    tmean = []
    cluster_i = get_data_cluster_i(data,i)
    eighteen_data = cluster_i.loc[cluster_i['year']==2018]
    nineteen_data = cluster_i.loc[cluster_i['year']==2019]
    for j in range(10):
        val = eighteen_data.loc[eighteen_data['month']==j+3]['tmean'].iloc[0]
        print("eighteen %f" % val)
        print(j+3)
        tmean.append(val)
    for j in range(7):
        val = nineteen_data.loc[nineteen_data['month']==j+1]['tmean'].iloc[0]
        print("nineteen %f" % val)
        print(j+1)        
        tmean.append(val)
    return tmean
```
